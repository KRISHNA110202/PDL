```python
import numpy as np
import matplotlib.pyplot as plt
from keras.applications import VGG16, ResNet50
from keras.applications.vgg16 import preprocess_input, decode_predictions
from keras.applications.resnet50 import preprocess_input as resnet_preprocess_input, decode_predictions as resnet_decode_predictions
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
```

```python
# Load a single image
def load_image(file_path, target_size):
    img = image.load_img(file_path, target_size=target_size)
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    return img

# Display image
def display_image(img_path):
    img = image.load_img(img_path)
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```

```python
# Part I: VGG16
def predict_with_vgg16(img_path):
    model = VGG16(weights='imagenet')
    target_size = (224, 224)  # VGG16 input size

    img = load_image(img_path, target_size)
    img = preprocess_input(img)

    predictions = model.predict(img)
    decoded_predictions = decode_predictions(predictions, top=10)

    for i, (_, label, prob) in enumerate(decoded_predictions[0]):
        print(f"Top-{i+1} Prediction: Class={label}, Name={label}, Probability={prob:.4f}")
```

```python
# Part II: ResNet50
def predict_with_resnet50(img_path):
    model = ResNet50(weights='imagenet')
    target_size = (224, 224)  # ResNet50 input size

    img = load_image(img_path, target_size)
    img = resnet_preprocess_input(img)

    predictions = model.predict(img)
    decoded_predictions = resnet_decode_predictions(predictions, top=10)

    for i, (_, label, prob) in enumerate(decoded_predictions[0]):
        print(f"Top-{i+1} Prediction: Class={label}, Name={label}, Probability={prob:.4f}")

```

```python
# Part III: Predict multiple images
def predict_multiple_images(image_paths):
    model = VGG16(weights='imagenet')
    target_size = (224, 224)

    for img_path in image_paths:
        img = load_image(img_path, target_size)
        img = preprocess_input(img)

        predictions = model.predict(img)
        decoded_predictions = decode_predictions(predictions, top=10)

        print(f"Predictions for Image: {img_path}")
        for i, (_, label, prob) in enumerate(decoded_predictions[0]):
            print(f"Top-{i+1} Prediction: Class={label}, Name={label}, Probability={prob:.4f}")
        print("\n")

```

```python
# Part IV: Data Augmentation
def data_augmentation_example(img_dir):
    datagen = ImageDataGenerator(
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest'
    )

    img = image.load_img(img_dir)
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)

    i = 0
    for batch in datagen.flow(x, batch_size=1):
        plt.figure(i)
        imgplot = plt.imshow(image.array_to_img(batch[0]))
        i += 1
        if i % 4 == 0:
            break

if __name__ == "__main__":
    # Single image prediction with VGG16 and ResNet50
    image_path = "C:\\Users\\shree\\Downloads\\239084.jpg"
    print("VGG16 Predictions:")
    predict_with_vgg16(image_path)
    print("\nResNet50 Predictions:")
    predict_with_resnet50(image_path)

    # Part IV: Data Augmentation Example
    img_dir = "C:\\Users\\shree\\Downloads\\239084.jpg"
    data_augmentation_example(img_dir)

    # Display the input image
    display_image(image_path)
```

```python
def visualize_multiple_images(image_paths):
    model = VGG16(weights='imagenet')
    target_size = (224, 224)

    # Data augmentation settings
    datagen = ImageDataGenerator(
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest'
    )

    for img_path in image_paths:
        # Load the original image
        original_img = load_image(img_path, target_size)
        original_img = preprocess_input(original_img)

        # Display the original image
        display_image(img_path)

        # Display predictions for the original image
        predictions = model.predict(original_img)
        decoded_predictions = decode_predictions(predictions, top=10)
        print(f"Predictions for Original Image: {img_path}")
        for i, (_, label, prob) in enumerate(decoded_predictions[0]):
            print(f"Top-{i+1} Prediction: Class={label}, Name={label}, Probability={prob:.4f}")
        print("\n")

        # Apply data augmentation and display augmented images
        augmented_images = datagen.flow(original_img, batch_size=4)
        for i, augmented_image in enumerate(augmented_images):
            if i >= 4:  # Display only the first 4 augmented images
                break
            augmented_image = preprocess_input(augmented_image)
            predictions = model.predict(augmented_image)
            decoded_predictions = decode_predictions(predictions, top=10)
            
            # Display the augmented image
            plt.figure()
            plt.imshow(image.array_to_img(augmented_image[0]))
            plt.axis('off')
            plt.show()

            # Display predictions for the augmented image
            print(f"Predictions for Augmented Image {i+1} of {img_path}")
            for j, (_, label, prob) in enumerate(decoded_predictions[0]):
                print(f"Top-{j+1} Prediction: Class={label}, Name={label}, Probability={prob:.4f}")
            print("\n")

if __name__ == "__main__":
    # Single image prediction with VGG16 and ResNet50
    image_path = "C:\\Users\\shree\\Downloads\\239084.jpg"
    print("VGG16 Predictions:")
    predict_with_vgg16(image_path)
    print("\nResNet50 Predictions:")
    predict_with_resnet50(image_path)

    # Part III: Predict and visualize multiple images with data augmentation
    image_paths = ["C:\\Users\\shree\\Downloads\\744461.jpg", "C:\\Users\\shree\\Downloads\\1219522.jpg", "C:\\Users\\shree\\Downloads\\621465 (1).jpeg"]
    visualize_multiple_images(image_paths)

    # Part IV: Data Augmentation Example
    img_dir = "C:\\Users\\shree\\Downloads\\239084.jpg"
    data_augmentation_example(img_dir)

    # Display the input image
    display_image(image_path)
```
