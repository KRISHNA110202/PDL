### Part-IV: Write a method normalizeRows()

```python
def normalizeRows(x):
    x_norm = np.linalg.norm(x, ord = 2, axis = 1, keepdims = True)
    x = x/x_norm
    return x
```

```python
x = np.array([[0, 3, 4],[1, 6, 4]])
print("normalizeRows(x) = " + str(normalizeRows(x)))
```

```python

# Create two NumPy arrays
x1 = np.array([7, 2, 5])
x2 = np.array([9, 2, 2])

# VECTORIZED ELEMENTWISE MULTIPLICATION
tic = time.process_time()
mul = np.multiply(x1, x2)
toc = time.process_time()
print("Elementwise multiplication =", mul)
print("----- Computation time =", 1000 * (toc - tic), "ms")

# VECTORIZED GENERAL DOT PRODUCT
tic = time.process_time()
dot = np.dot(x2, x1)
toc = time.process_time()
print("Generalized dot product =", dot)
print("----- Computation time =", 1000 * (toc - tic), "ms")
```

### Part-VI: Implement L1 and L2 loss functions

```python
# 1.
# GRADED FUNCTION: L1
def L1(yhat, y):
    loss = np.sum(abs(y - yhat), axis = 0, keepdims = True)
    return loss
y = np.array([1, 0, 0, 1, 1])
yhat = np.array([.9, 0.2, 0.1, .4, .9])
print("L1 = " + str(L1(yhat,y)))
```

```python
# 2.
# GRADED FUNCTION: L2
def L2(yhat, y):
    loss = np.dot((y - yhat),(y - yhat))
    return loss
y = np.array([1, 0, 0, 1, 1])
yhat = np.array([.9, 0.2, 0.1, .4, .9])
print("L2 = " + str(L2(yhat,y)))
```
